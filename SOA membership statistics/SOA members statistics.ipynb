```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import membership csv numbers
df = pd.read_csv('membership_numbers.csv')
```

```python
# Graph for SOA members by year
plt.figure(figsize = (14,10))
sns.set(font_scale = 1.15)
plt.title('SOA members by Year', size = 12, font = 'Georgia', color = 'black')
sns.barplot(data = df, x = 'Year', y = 'Members', hue = 'Certification',
            ci = None)
plt.xlabel('Year', fontsize = 11, font = 'Georgia')
plt.ylabel('Number of members', fontsize = 11, font = 'Georgia')
```

```python
# Graph for SOA members by location
plt.figure(figsize = (14,10))
sns.set(font_scale = 1.15)
sns.barplot(data = df, x = 'Year', y = 'Members', hue = 'Country', ci = None)
plt.title('SOA members by Location', size = 14, font = 'Georgia', color = 'black')
plt.xlabel('Year', fontsize = 11, font = 'Georgia')
plt.ylabel('Number of members', fontsize = 11, font = 'Georgia')
```
